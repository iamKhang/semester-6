Chiều cao của thanh niên ở Việt Nam (2025) được mô tả bằng phân phối chuẩn N (muy = 172 cm, sigma = 5 cm)

```python
# a) Xác suất chọn một bạn ngẫu nhiên gọi là lùn (nhỏ hơn 162cm)
from scipy.stats import norm

mu = 172  # Trung bình
sigma = 5  # Độ lệch chuẩn

# Tính xác suất cho chiều cao nhỏ hơn 162 cm
prob = norm.cdf(162, mu, sigma)

print(prob)
```

```python
# Giả sử, thanh niên nam được xem là cao lớn (> 182 (cm)). Hỏi xác suất chọn ngẫu nhiên
# một thanh niên nam cao lớn là bao nhiêu?
mu = 172  # Trung bình
sigma = 5  # Độ lệch chuẩn

# Tính xác suất cho chiều cao lớn hơn 182 cm
prob = norm.sf(182, mu, sigma)

print(prob)
```

```python
# Tính xác suất cho chiều cao trong khoảng từ 168 cm đến 182 cm
prob = norm.cdf(182, mu, sigma) - norm.cdf(168, mu, sigma)

print(prob)
```

```python
# Cho biết 6% là thấp bé, vậy thế nào là chuẩn thấp bé của thanh niên nam (cm)
mu = 172  # Trung bình
sigma = 5  # Độ lệch chuẩn

# Tìm chiều cao mà tại đó chỉ có 6% thanh niên nam có chiều cao thấp hơn
height = norm.ppf(0.06, mu, sigma)

print(height)
```

```python
# Cho biết đến 9% cao lớn, vậy cao lớn có ngưỡng chiều cao là bao nhiêu (cm)?
# Tìm ngưỡng chiều cao mà ở đó chỉ có 9% thanh niên nam cao hơn
height_threshold = norm.ppf(0.91, mu, sigma)

print(height_threshold)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 172  # Trung bình
sigma = 5  # Độ lệch chuẩn

# Tạo một mảng các giá trị x từ -6 đến 6 với bước là 0.25
x = np.arange(-6, 6, 0.25)

# Tính pmf và cdf cho mỗi giá trị x
pmf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

# Vẽ biểu đồ pmf
# plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, pmf, label='PMF')
plt.title('Probability Mass Function')
plt.xlabel('Height (cm)')
plt.ylabel('Probability')

# Vẽ biểu đồ cdf
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF')
plt.title('Cumulative Distribution Function')
plt.xlabel('Height (cm)')
plt.ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```
