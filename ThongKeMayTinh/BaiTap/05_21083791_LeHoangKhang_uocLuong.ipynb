```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
# Tạo ngẫu nhiên 10000 phần tử có chiều cao từ 120 đến 200cm
POP = np.random.uniform(120, 200, 10000)

# Tính chiều cao trung bình (Kỳ vọng) và độ lệch chuẩn của quần thể
pop_mean = np.mean(POP)
print(f'Chiều cao trung bình của quần thể là: {pop_mean} cm')
pop_std = np.std(POP)
print(f'Độ lệch chuẩn của quần thể là: {pop_std} cm')
# b. Tính tỉ lệ người cao trong quần thể.
tall_rate = np.sum(POP > 180)/len(POP)
print(f'Tỉ lệ người cao trong quần thể là {tall_rate} người')
# c. Vẽ histogram chiều cao của quần thể
plt.hist(POP, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')
plt.title('Population height distribution')
plt.xlabel('Height (cm)')
```

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

```python
# Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)
sample_variances = []
for _ in range(1000):
    sample = np.random.choice(POP, 100)
    sample_variance = np.var(sample)
    sample_variances.append(sample_variance)

plt.hist(sample_variances, bins=30, density=True, alpha=0.6, color='b', edgecolor='black')
plt.title('Histogram of Sample Variances')
plt.xlabel('Sample Variance')
plt.ylabel('Density')
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng 

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
import numpy as np
from scipy.stats import norm, t

# Mẫu cân nặng của gà
weights = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])

# Tính trung bình mẫu và kích thước mẫu
mean = np.mean(weights)
size = len(weights)

# a. Biết sigma = 0.3
sigma = 0.3
confidence_interval_a = norm.interval(0.95, loc=mean, scale=sigma/np.sqrt(size))

# b. Không biết sigma
std_error = np.std(weights, ddof=1) / np.sqrt(size)
confidence_interval_b = t.interval(0.95, df=size-1, loc=mean, scale=std_error)

print("Khoảng tin cậy 95% khi biết sigma: ", confidence_interval_a)
print("Khoảng tin cậy 95% khi không biết sigma: ", confidence_interval_b)
```
